```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Importing the dataset
dataset = pd.read_csv("Position_Salaries.csv")

dataset.head()
```

```python
x, y = dataset.loc[:, "Level"].values.reshape(len(dataset), 1), dataset.loc[:, "Salary"].values
```

```python
plt.figure(figsize = (8, 5))

plt.xlabel("Position level", fontsize = 15)
plt.ylabel("Salary", fontsize = 15)
plt.title("The relation between level and salary", fontsize = 20)

plt.plot(x, y, marker = 'x', markersize = 10, linestyle = '', color = "red")
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor = linear_regressor.fit(x, y)

ypredicted = linear_regressor.predict(x)
print("The Accuracy = {0:0.2f}%".format(linear_regressor.score(x, y) * 100))
```

```python
# Visualising the Linear Regression results
plt.figure(figsize = (8, 5))
plt.xlabel("Position level", fontsize = 15)
plt.ylabel("Salary", fontsize = 15)
plt.title("Truth or Bluff (Linear Regression)", fontsize = 20)

plt.plot(x, y, marker = 'x', markersize = 10, linestyle = '', color = "red")
plt.plot(x, ypredicted, linestyle = '-.', color = "blue", linewidth = 1)
```

## Fitting Polynomial Regression to the dataset

```python
from sklearn.preprocessing import PolynomialFeatures

# Changing a degree of our curve gives more visualization for the best fitting curve
poly_transformer = PolynomialFeatures(degree = 4)
polyx = poly_transformer.fit_transform(x)

poly_regressor = LinearRegression()
poly_regressor = poly_regressor.fit(polyx, y)

ypredicted = poly_regressor.predict(polyx)
print("The Accuracy = {0:0.2f}%".format(poly_regressor.score(polyx, y) * 100))
```

```python
# Visualising the Polynomial Regression results (for higher resolution and smoother curve)
plt.figure(figsize = (8, 5))

plt.xlabel("Position level", fontsize = 15)
plt.ylabel("Salary", fontsize = 15)
plt.title("Truth or Bluff (Polynomial Regression)", fontsize = 20)

plt.plot(x, y, marker = 'x', markersize = 10, linestyle = '', color = "red")
plt.plot(x, ypredicted, linestyle = '-.', color = "blue", linewidth = 1)
```
